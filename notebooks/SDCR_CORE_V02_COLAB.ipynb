### **4. Verify Manifest and ZIP Contents**
We verify that all generated outputs match the SHA-256 manifest and that the ZIP file contains all required assets.

```python
import json
import zipfile
from sdcr_core.io.manifest import compute_sha256

# 1. Verify Manifest Checksums
print("=== Manifest Verification ===")
with open("results/SHA256_MANIFEST.json", "r") as f:
    manifest = json.load(f)

for filepath, info in manifest.items():
    actual_path = filepath
    if filepath == "SDCR_CORE_V02_RESULT_PACKAGE.zip":
        actual_path = filepath
    elif filepath.startswith("figures/"):
        actual_path = filepath
    else:
        actual_path = os.path.join("results", filepath)
        
    if os.path.exists(actual_path):
        h = compute_sha256(actual_path)
        status = "PASS" if h == info["sha256"] else "FAIL"
        print(f"{filepath}: {status} (size: {info['size_bytes']} bytes)")
    else:
        print(f"{filepath}: NOT FOUND")

# 2. Verify ZIP Contents
print("\n=== ZIP Contents Verification ===")
zip_name = "SDCR_CORE_V02_RESULT_PACKAGE.zip"
if os.path.exists(zip_name):
    with zipfile.ZipFile(zip_name, "r") as z:
        print(f"ZIP file '{zip_name}' contains:")
        for name in z.namelist():
            print(f"  - {name}")
else:
    print(f"{zip_name} NOT FOUND")
```
